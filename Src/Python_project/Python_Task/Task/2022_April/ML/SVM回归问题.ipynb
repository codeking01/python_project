```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib.inline
from sklearn.svm import SVR
```

```python
#0到3，然后分成50份， 一维数据
'''
    reshape(1,-1)转化成1行：
    reshape(2,-1)转换成两行：
    reshape(-1,1)转换成1列：
    reshape(-1,2)转化成两列'''
X=np.linspace(0,3*np.pi,50).reshape(-1,1)
y=np.sin(X)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.svm import SVR
# 线性
svr_linear= SVR(kernel='linear')
# 高斯核函数RBF
svr_rbf= SVR(kernel='rbf')
# 多项式
svr_poly= SVR(kernel='poly')
# 训练模型
svr_linear.fit(X,y.ravel())
svr_rbf.fit(X,y.ravel())
svr_poly.fit(X,y.ravel())
```

```python
X_test=np.linspace(0,3*np.pi,128).reshape(-1,1)
```

```python
y1=svr_linear.predict(X_test)
plt.scatter(X_test,y1)
y2=svr_rbf.predict(X_test)
plt.scatter(X_test,y2)
y3=svr_poly.predict(X_test)
plt.scatter(X_test,y3)
```

```python
plt.scatter(X,y)
plt.plot(X_test,y1,c='blue')
plt.plot(X_test,y2,c="pink")
plt.plot(X_test,y3,c="orange")
plt.legend(["orange_data","Linear","Rbf","poly"])
```

```python
# 计算预测的分数
from sklearn.metrics import explained_variance_score
y_test=np.sin(X_test)
score=explained_variance_score(y_test,y2)
print('预测准确率evs: %.2f%%'%(score*100.0))
```
